```python
import pandas as pd
import sqlite3
```

```python
conexao = sqlite3.connect('Netflix.db')
consulta_sql = "SELECT * FROM Data"
df = pd.read_sql_query(consulta_sql, conexao)
conexao.close()

df
```

```python

print(df['director'].isnull().sum())
```

```python

import random
for director,line in df.iterrows():
    if pd.isnull(line['director']):
        samples = df[df['country'] == line['country']]
        samples = samples.dropna(subset=['director'])
        if not samples.empty:
            diretor = samples.sample()['director']
            df.at[director,'director'] = diretor.values[0]
 
```

```python
import random
for director,line in df.iterrows():
    if pd.isnull(line['director']):
        samples = df[df['cast'] == line['cast']]
        samples = samples.dropna(subset=['director'])
        if not samples.empty:
            diretor = samples.sample()['director']
            df.at[director,'director'] = diretor.values[0]
 
```

```python
for country,line in df.iterrows():
    if pd.isnull(line['country']):
        samples = df[df['director'] == line['director']]
        samples = samples.dropna(subset=['country'])
        if not samples.empty:
            pais = samples.sample()['country']
            df.at[country,'country'] = pais.values[0]

```

```python
print(df['cast'].isnull().sum())
```

```python

for index, row in df.iterrows():
    if pd.isnull(row['cast']):
        director = row['director']
        samples = df[(df['director'] == director) & ~df['cast'].isnull()]
        if not samples.empty:
            cast_value = samples.sample()['cast'].values[0]
            df.at[index, 'cast'] = cast_value
```
